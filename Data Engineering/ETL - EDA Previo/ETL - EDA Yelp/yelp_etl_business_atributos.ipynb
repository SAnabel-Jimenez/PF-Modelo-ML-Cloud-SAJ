# YELP - BUSINESS

```python
import pandas as pd
from pandas import json_normalize
```

```python
def nulos_df(df):
    cant_nulos = df.isnull().sum()
    porcentaje_nulos = round((cant_nulos/len(df))*100, 2)
    resumen =pd.DataFrame({'CantNulos':cant_nulos,'Nulos(%)': porcentaje_nulos})

    return resumen
```

```python
Yelp_business = pd.read_pickle("Yelp/business.pkl")
print(Yelp_business.shape)
Yelp_business.head(3)
```

```python
Yelp_business.columns
```

```python
Yelp_business.info()
```
